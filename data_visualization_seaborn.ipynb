```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
# scatter plot using seaborn
sns.scatterplot(x="total_bill",y="tip",data=tips_df)
plt.title("Scatter Plot")
plt.show()
```

```python
# line plot

sns.lineplot(x="size",y="total_bill",data=tips_df)
plt.title("Line Plot")
plt.show()
```

```python
# categorical plot

# bar plot
sns.barplot(x="day",y="total_bill",data=tips_df)
plt.title("Bar plot")
plt.show()
```

```python
# box plot

sns.boxplot(x="day",y="total_bill",data=tips_df)
plt.title("Box Plot")
plt.show()
```

```python
#violinplot
sns.violinplot(x="day",y="total_bill",data=tips_df)
plt.title("Violin Plot")
plt.show()
```

```python
# histogram

sns.histplot(tips_df["total_bill"],bins=10,kde=True)
plt.title("Histogram")
plt.show()
```

```python
# kde plot

sns.kdeplot(tips_df["total_bill"],fill=True)
plt.title("KDE plot")
plt.show()
```

```python
#pairplots
sns.pairplot(tips_df)
```

```python
# corelations

tips_df.info()
```

```python
corr = tips_df[["total_bill","tip","size"]].corr()
corr
```

```python
#heatmap

sns.heatmap(corr,annot=True,cmap='coolwarm')
```
